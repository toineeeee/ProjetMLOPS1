```python
import yfinance as yf
import pandas as pd

bank_tickers = [
    "JPM", "BAC", "WFC", "C", "GS", "MS", "USB", "PNC", "TFC", "COF",
    "RF", "FITB", "STT", "MTB", "BK", "KEY", "CFG", "HBAN", "ALLY", "CMA",
    "BPOP", "FHN", "ZION", "SNV", "FRC", "OZK", "EWBC", "PB", "BOH", "WAL"
]

# recueil de données macroéconomiques qui sont fortement corrélés aux variation de cours de bourse des banques
macro_tickers = {
    "^GSPC": "S&P_500",
    "^DJI": "Dow_Jones",
    "^IXIC": "Nasdaq",
    "^VIX": "VIX",  # (volatilité)
    "^TNX": "US_10Y_Treasury",  # taux des obligations US 10 ans
    "^IRX": "US_3M_Treasury"  # taux des obligations US 3 mois (proche des taux directeurs de la Fed)
}

date_start = "2020-01-01"
date_end = "2023-12-31"

df_prices = pd.DataFrame()

# Téléchargement des données boursières des banques et indices
for ticker in bank_tickers + list(macro_tickers.keys()):
    print(f"Téléchargement des données pour {ticker}...")
    data = yf.download(ticker, start=date_start, end=date_end)[["Close"]]
    
    if data.empty:
        print(f"Aucune donnée disponible pour {ticker}, next ")
        continue
    
    # Renommer la colonne
    column_name = macro_tickers.get(ticker, ticker)  
    data.rename(columns={"Close": column_name}, inplace=True)
    
    # Fusionner avec le DataFrame principal
    if df_prices.empty:
        df_prices = data
    else:
        df_prices = df_prices.join(data, how="outer")

# nettoyage et suppression des valeurs manquantes
df_prices.dropna(inplace=True)

# print(df_prices)
print(df_prices.head())

df_prices.to_csv("bank_stock_prices_with_macro_2020_2023.csv")
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

#if isinstance(df_prices.columns, pd.MultiIndex):
#    df_prices.columns = [col[0] if isinstance(col, tuple) else col for col in df_prices.columns]

# nettoyage noms des colonnes
df_prices.columns = df_prices.columns.str.strip()

# sélection des variables explicatives
features = [col for col in df_prices.columns if col != "JPM"]

# normalisation des données
scaler = MinMaxScaler(feature_range=(0, 1))
df_scaled = scaler.fit_transform(df_prices)

# DataFrame normalisé
df_scaled = pd.DataFrame(df_scaled, columns=df_prices.columns, index=df_prices.index)

# définition séquences pour LSTM, sur 30 jours d'hisotrique et prévision sur les 7 prochains jours
def create_sequences(data, target_column, seq_length=30, pred_horizon=7):
    X, y = [], []
    for i in range(len(data) - seq_length - pred_horizon):
        X.append(data.iloc[i:i+seq_length].values) 
        y.append(data.iloc[i+seq_length:i+seq_length+pred_horizon][target_column].values) 
    return np.array(X), np.array(y)

# création des séquences pour LSTM
seq_length, pred_horizon = 30, 7
X, y = create_sequences(df_scaled, "JPM", seq_length, pred_horizon)

# division des données pour attribuer 80% à l'entrainement (le reste sera pour le test)
train_size = int(0.8 * len(X))
X_train, X_test, y_train, y_test = X[:train_size], X[train_size:], y[:train_size], y[train_size:]

```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# définition du modèle LSTM
model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dropout(0.2),
    LSTM(50, return_sequences=False),
    Dropout(0.2),
    Dense(25),
    Dense(7)  # 7 sorties pour prédire les 7 jours suivants
])

# compilation du modèle
model.compile(optimizer="adam", loss="mean_squared_error")

# Entraînement du modèle
history = model.fit(X_train, y_train, epochs=50, batch_size=16, validation_data=(X_test, y_test))

# sauvegarde du modèle entraîné
model.save("lstm_model_jpm.h5")

# Affichage des résultats de l'entraînement
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label="Loss Entraînement")
plt.plot(history.history['val_loss'], label="Loss Validation")
plt.xlabel("Épochs")
plt.ylabel("Erreur")
plt.title("Évolution de la loss pendant l'entraînement")
plt.legend()
plt.grid()
plt.show()
```

```python
from tensorflow.keras.models import load_model

model = load_model("lstm_model_jpm.h5")

# sélection de la dernière séquence de test pour prédire les 7 jours suivants
last_sequence = X_test[-1]
last_sequence = last_sequence.reshape(1, seq_length, X_train.shape[2])

predicted_prices = model.predict(last_sequence)[0]  # Sortie sous forme de tableau 1D

# inversion la normalisation, pour revenir aux données d'origine (à afficher sur le graph)
predicted_prices = scaler.inverse_transform(
    np.concatenate((np.zeros((7, len(features))), predicted_prices.reshape(-1, 1)), axis=1)
)[:, -1]

# récupération des vraies valeurs pour comparer
true_prices = scaler.inverse_transform(
    np.concatenate((np.zeros((7, len(features))), y_test[-1].reshape(-1, 1)), axis=1)
)[:, -1]


future_dates = pd.date_range(start=df_prices.index[-1], periods=8, freq='D')[1:]

# affichage des résultats
plt.figure(figsize=(12, 6))
plt.plot(future_dates, true_prices, label="Prix Réels", marker='o', color='blue')
plt.plot(future_dates, predicted_prices, label="Prédictions LSTM", marker='x', color='red', linestyle="dashed")
plt.xlabel("Date")
plt.ylabel("Prix de clôture ($)")
plt.title("Prédiction des prix de JPM sur les 7 prochains jours")
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# définir la cible et les features pour MS
target = "MS"
features = [col for col in df_prices.columns if col != target]

X = df_prices[features]
y = df_prices[target]

# mêmes paramètres que pour le LSTM
seq_length = 30  
pred_horizon = 7 

X_rf, y_rf = [], []
for i in range(len(X) - seq_length - pred_horizon):
    X_rf.append(X.iloc[i:i+seq_length].values.flatten())  # Transformer en un seul vecteur
    y_rf.append(y.iloc[i+seq_length:i+seq_length+pred_horizon].values)  # Prédire 7 jours

X_rf, y_rf = np.array(X_rf), np.array(y_rf)

# entrainement/test : séparation des données
X_train_rf, X_test_rf, y_train_rf, y_test_rf = train_test_split(X_rf, y_rf, test_size=0.2, shuffle=False)

# création du modèle Random Forest
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# entraînement du modèle
rf_model.fit(X_train_rf, y_train_rf)

# prédictions sur les données de test
y_pred_rf = rf_model.predict(X_test_rf)

# évaluation du modèle
mae_rf = mean_absolute_error(y_test_rf, y_pred_rf) #évaluation de la moyenne des erreurs
rmse_rf = np.sqrt(mean_squared_error(y_test_rf, y_pred_rf)) #erreur quadratique

print(f"MAE (Random Forest) : {mae_rf:.2f}")
print(f"RMSE (Random Forest) : {rmse_rf:.2f}")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Vérifier que les variables existent
if "X_test_rf" not in globals() or "rf_model" not in globals():
    raise ValueError("Erreur : Les variables nécessaires ne sont pas définies. Vérifie l'exécution des étapes précédentes.")

# Sélectionner la dernière séquence pour prédire les 7 jours suivants
last_sequence_rf = X_test_rf[-1].reshape(1, -1)

# Faire une prédiction sur 7 jours
predicted_rf = rf_model.predict(last_sequence_rf)[0]

# Récupérer les vraies valeurs pour comparaison
true_values_rf = y_test_rf[-1]

# Générer les dates pour l'affichage
future_dates_rf = pd.date_range(start=df_prices.index[-1], periods=8, freq='D')[1:]

# Création du graphique
plt.figure(figsize=(12, 6))

# Tracer les prix réels en bleu avec des marqueurs cercles
plt.plot(future_dates_rf, true_values_rf, label="Prix Réels", marker='o', color='blue', linestyle="-")

# Tracer les prédictions Random Forest en rouge pointillé avec marqueur x
plt.plot(future_dates_rf, predicted_rf, label="Prédictions Random Forest", marker='x', color='red', linestyle="dashed")

# Labels et titre
plt.xlabel("Date")
plt.ylabel("Prix de clôture ($)")
plt.title("Prédiction des prix de MS sur les 7 prochains jours (Random Forest)")

# Légende et grille
plt.legend()
plt.grid()

# Affichage du graphique
plt.show()
```

```python
# calcul des résidus
residuals_jpm = predicted_prices - true_prices
residuals_ms = predicted_rf - true_values_rf

# histogramme commun
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.hist([residuals_jpm, residuals_ms], bins=10, color=['blue', 'red'], 
         label=['Résidus JPM (LSTM)', 'Résidus MS (Random Forest)'], alpha=0.7)
plt.xlabel("Résidus (Prédiction - Réel)")
plt.ylabel("Fréquence")
plt.title("Comparaison des résidus pour JPM et MS")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# calcul du prix moyen pour JPM et MS
mean_price_jpm = df_prices["JPM"].mean()
mean_price_ms = df_prices["MS"].mean()
print(f"moy prix JPM : {mean_price_jpm:.2f}")
print(f"moy prix MS  : {mean_price_ms:.2f}")

# calcul des résidus pour JPM (modèle LSTM) et pour MS (Random Forest)
residuals_jpm = predicted_prices - true_prices
residuals_ms = predicted_rf - true_values_rf

# passage des résidus en pourcentage
residuals_pct_jpm = (residuals_jpm / true_prices) * 100
residuals_pct_ms = (residuals_ms / true_values_rf) * 100

# histogramme

plt.figure(figsize=(10, 6))
plt.hist([residuals_pct_jpm, residuals_pct_ms], bins=10, color=['blue', 'red'],
         label=['Résidus % JPM (LSTM)', 'Résidus % MS (Random Forest)'], alpha=0.7)
plt.xlabel("Résidus en pourcentage (%)")
plt.ylabel("Fréquence")
plt.title("Histogramme des résidus en pourcentage pour JPM et MS")
plt.legend()
plt.grid(True)
plt.show()
```
